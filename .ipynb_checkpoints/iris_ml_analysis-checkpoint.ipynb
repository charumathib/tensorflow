```python
# import required libraries
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
```

```python
# load the data
data = pd.read_csv("./iris.csv")
```

```python
# shuffle data
data = data.sample(frac=1).reset_index(drop=True)
cols = data.columns
flower_data = []
flower_labels = []

for i in range (0, 150):
    row = 0
    temp = []
    for col in cols:
        #parse data into list of measurements and corresponding labels
        if(row != 4):
            temp.append(data[col][i])
        else:
            if(data[col][i] == 'Iris-setosa'):
                flower_labels.append(0)
            elif(data[col][i] == 'Iris-versicolor'):
                flower_labels.append(1)
            else:
                flower_labels.append(2)
        row += 1
    flower_data.append([temp])
```

```python
# create training and testing sets
x_train, x_test, y_train, y_test = train_test_split(flower_data, flower_labels, test_size=0.3)

#train and deploy the model
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(512, activation=tf.nn.relu),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100)
model.evaluate(x_test, y_test)
```
